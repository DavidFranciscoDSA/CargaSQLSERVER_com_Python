```python
driver_ODBC = 'ODBC Driver 17 for SQL Server'
nome_servidor = 'DAVIDFRANCISCO\SQLEXPRESS'
nome_BD = 'AulaPython'
tabelaSQL = '[Clientes]'
usuario = 'dsds'
senha = '1234'
```

Caso a coluna tenha catacteres especiais como ´~@ ex: nome, email, fone é preciso converter - Usa na linha de baixo do for

```python
'''
    linha.email = str(linha.email)  # Converter para o tipo 'str' antes da inserção
    linha.country = str(linha.country)
    linha.state = str(linha.state)
    linha.street = str(linha.street)
    linha.number = str(linha.number)
    linha.additionals = str(linha.additionals) 
'''
```

```python
# conexão do banco de dados SQL SERVER sem senha
conexaoDB = pyodbc.connect(f'DRIVER={driver_ODBC};'
                           f'SERVER={nome_servidor};'
                           f'DATABASE={nome_BD};'
                           'Trusted_Connection=yes;')

cursor = conexaoDB.cursor()   # criando cursor de comando
    
#Apaga caso haja uma tabela com esse nome

cursor.execute('truncate table '+tabelaSQL) 
for index, linha in df.iterrows():
    linha.email = str(linha.email)  # Converter para o tipo 'str' antes da inserção
    linha.country = str(linha.country)
    linha.state = str(linha.state)
    linha.street = str(linha.street)
    linha.number = str(linha.number)
    linha.additionals = str(linha.additionals) 
    cursor.execute("Insert into "+tabelaSQL+"(id,created_at,first_name,last_name,email,cell_phone,country,state,street,number,additionals)values(?,?,?,?,?,?,?,?,?,?,?)",
                   linha.id,linha.created_at,linha.first_name,linha.last_name,linha.email,linha.cell_phone,linha.country,linha.state,linha.street,linha.number,linha.additionals,) 
     # inserir colunas e quantas colunas tiver passar quantidade de ??
cursor.commit()   # validar dados no SQL Server
cursor.close()    #Fechar Cursor
conexaoDB.close() #Fechar Conexao
```
